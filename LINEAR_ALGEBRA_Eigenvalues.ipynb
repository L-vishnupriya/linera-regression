```python
import numpy as np
import matplotlib.pyplot as plt

# Define the matrix A
A = np.array([[1, 6],
              [5, 2]])
```

```python
# Define the vectors U and V
U = np.array([[6],
              [-5]])
V = np.array([[3],
              [-2]])
```

```python
# Compute eigenvalues and eigenvectors of A
eigenvalues, eigenvectors = np.linalg.eig(A)

# Plotting
plt.figure(figsize=(8, 6))
```

```python
# Plotting the vectors
plt.quiver(0, 0, U[0], U[1], angles='xy', scale_units='xy', scale=1, color='r', label='U')
plt.quiver(0, 0, V[0], V[1], angles='xy', scale_units='xy', scale=1, color='b', label='V')

# Plotting eigenvectors
for i in range(len(eigenvectors)):
    plt.quiver(0, 0, eigenvectors[i][0], eigenvectors[i][1], angles='xy', scale_units='xy', scale=1, color='g',
               label=f'eigenvector {i+1}')

plt.xlim(-7, 7)
plt.ylim(-7, 7)
plt.xlabel('X')
plt.ylabel('Y')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.legend()
plt.title('Visualization of Eigenvectors')
plt.show()

# Check if U and V are eigenvectors
for i in range(len(eigenvalues)):
    if np.allclose(A @ U, eigenvalues[i] * U):
        print("U is an eigenvector of A with eigenvalue:", eigenvalues[i])
    if np.allclose(A @ V, eigenvalues[i] * V):
        print("V is an eigenvector of A with eigenvalue:", eigenvalues[i])
```
